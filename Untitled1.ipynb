```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv('/home/jp/dataset/Used_Bikes.csv')
```

```python
df.head()
```

```python
l=df.groupby('city').agg(count1=('city','count')).sort_values(by='count1',ascending=False).head().reset_index()
l
plt.figure(figsize=(13,6))
plt.bar(x=l['city'],height=l['count1'],color=['red','blue','yellow','green','violet'])

plt.title('TOP FIVE BIKE SALE CITY',fontsize=16,color='black')
plt.xticks(rotation=90)
```

```python
ownergroupby=df.groupby('owner').agg(count=('owner','count'),totalSales=('price','sum')).reset_index()
plt.pie(x=ownergroupby['totalSales'],autopct='%1.1f%%',startangle=90,labels=['first owner','fourth','third','second'])
```

```python
Scalr=StandardScaler()
df['age']=round(df['age'],2)
df
```

```python
x=df[['owner','city','power','brand']]

x
y=df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
x
for col in x.columns:
    x[col]=lab.fit_transform(x[col])
x
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42
                                              )
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred,squared=True)  # RMSE
r2 = r2_score(y_test, y_pred)
```

```python
print(f'the value of mean_square{mse}')
print(f'rmse value is {rmse}')
print(f'the r2 is {r2}')
```
